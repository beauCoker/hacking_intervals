```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
import sklearn.metrics
from scipy import optimize
import numpy.linalg as la
%matplotlib inline
```

Simulate data

```python
n = 200
npr.seed(123)

X = npr.uniform(0,10, (n,2))

def rand_outcome(s):
    return npr.normal(s[1]-s[0],scale=3)

Y = np.apply_along_axis(rand_outcome,1,X)

xnew = np.array([5,5])
```

```python
def distance(z1,z2,M):
    # Mahalanobis distance
    return (z2-z1).T @ M @ (z2-z1)
```

```python
def plot_distance(M, X, Y, C, xnew, ngrid=20):
    
    XS, YS = np.meshgrid(np.linspace(0,10,ngrid), np.linspace(0,10,ngrid))

    dist = np.array(list(map(lambda i: distance(xnew,np.array([XS.ravel()[i],YS.ravel()[i]]),M), range(XS.size)))).reshape(XS.shape)

    fig, ax = plt.subplots(figsize=(6,5))
    cax=ax.scatter(X[:,0],X[:,1], c=Y)
    ax.contour(XS, YS, dist, levels=np.array([C]))
    fig.colorbar(cax)
    
```

Find the minimum loss solution ('MLE') 

```python
cutoff = 1e-34
def kernel(z1,z2,M):
    return np.exp(-distance(z1,z2,M))
    #return np.sqrt((z2-z1).T @ np.identity(2) @ (z2-z1)) # Euclidean


def kernel_loss(A, X, Y):
    cutoff = 1e-34
    
    # Compute Mahalanobis matrix
    M = A.reshape((2,2)).T @ A.reshape((2,2))
    
    # Distances
    K = sklearn.metrics.pairwise.pairwise_distances(X, metric=kernel,M=M)
    K[K<cutoff] = 0
    
    # Regression prediction
    KY = K * Y[np.newaxis,:]
    KY_sum = np.sum(KY, axis=1) - np.diag(KY) #Numerator
    K_sum = np.sum(K,axis=1)-np.diag(K) # Denominator

    
    if (sum(K_sum==0)):
        #print("Found zero")
        #print(np.sum(K[K_sum==0,:]))
        #print(K_sum)
        #print(K[K_sum==0,:])
        #print(K)
        pass
    
    Yhat = KY_sum / K_sum
    Yhat[np.isnan(Yhat)]=0
    return np.sum((Y - Yhat)**2)
```

```python
A0 = np.identity(2)

res_mle = optimize.minimize(kernel_loss, 
                        A0.ravel(), 
                        args=(X, Y), 
                        method='Nelder-Mead')
```

```python
A_mle = res_mle.x.reshape((2,2))
M_mle = A_mle.T @ A_mle
```

Now find the hacking interval.

```python
def kernel_pred(A, X, Y, C, xnew):
    cutoff = 1e-34

    # Compute Mahalanobis matrix
    M = A.reshape((2,2)).T @ A.reshape((2,2))

    # Distances
    K = sklearn.metrics.pairwise.pairwise_distances(X, xnew[np.newaxis,:], metric=kernel,M=M).T
    K[K<cutoff] = 0
    
    # Regression prediction
    KY = K * Y[np.newaxis,:]
    KY_sum = np.sum(KY, axis=1) # Numerator
    K_sum = np.sum(K,axis=1) # Denominator

    Yhat = KY_sum / K_sum
    Yhat[np.isnan(Yhat)]=0
    return np.asscalar(Yhat)

def kernel_pred_min(A, X, Y, C, xnew):
    return kernel_pred(A, X, Y, C, xnew)
    
def kernel_pred_max(A, X, Y, C, xnew):
    return -kernel_pred(A, X, Y, C, xnew)
```

```python
# Prediction for MLE
kernel_pred(A_mle, X, Y, [], xnew)
```

```python
A0 = np.identity(2)
C = 2000

def kernel_loss_cons(A, X, Y, C, xnew):
    return C - kernel_loss(A, X, Y)

cons = [{'type':'ineq', 'fun': kernel_loss_cons, 'args': (X, Y, C, xnew)}]
```

```python
res_min = optimize.minimize(kernel_pred_min, 
                        A0.ravel(), 
                        args=(X, Y, C, xnew), 
                        constraints=cons,
                        method='SLSQP')
```

```python
res_max = optimize.minimize(kernel_pred_max, 
                        A0.ravel(), 
                        args=(X, Y, C, xnew), 
                        constraints=cons,
                        method='SLSQP')
```

```python
A_min = res_min.x.reshape((2,2))
M_min = A_min.T @ A_min
```

```python
A_max = res_max.x.reshape((2,2))
M_max = A_max.T @ A_max
```

```python
# Check the constraint (should be positive)
print(kernel_loss_cons(A_min, X, Y, C, xnew))
print(kernel_loss_cons(A_max, X, Y, C, xnew))
```

```python
### Predictions of xnew

# Prediction for MLE
print(kernel_pred(A_mle, X, Y, [], xnew))

# Min prediction
print(kernel_pred(A_min, X, Y, [], xnew))

# Max prediction
print(kernel_pred(A_max, X, Y, [], xnew))
```

```python
### Plot results

plt.rcParams.update({'font.size': 16})

XS, YS = np.meshgrid(np.linspace(0,10,200), np.linspace(0,10,200))

# Side-by-side
dist1 = np.array(list(map(lambda i: distance(xnew,np.array([XS.ravel()[i],YS.ravel()[i]]),M_mle), range(XS.size)))).reshape(XS.shape)
dist2 = np.array(list(map(lambda i: distance(xnew,np.array([XS.ravel()[i],YS.ravel()[i]]),M_min), range(XS.size)))).reshape(XS.shape)
dist3 = np.array(list(map(lambda i: distance(xnew,np.array([XS.ravel()[i],YS.ravel()[i]]),M_max), range(XS.size)))).reshape(XS.shape)

fig, ax = plt.subplots(1,3,figsize=(15,5))
ax[0].scatter(X[:,0],X[:,1], c=Y)
ax[0].contour(XS, YS, dist2, levels=np.array([1]))
ax[0].set_title("Minimum prediction")
ax[0].set_aspect('equal', 'box')
ax[0].set_xticks(np.arange(0,12,2))

ax[1].scatter(X[:,0],X[:,1], c=Y)
ax[1].contour(XS, YS, dist1, levels=np.array([1]))
ax[1].set_title("Minimum loss")
ax[1].set_aspect('equal', 'box')
ax[1].set_xticks(np.arange(0,12,2))

cax = ax[2].scatter(X[:,0],X[:,1], c=Y)
ax[2].contour(XS, YS, dist3, levels=np.array([1]))
ax[2].set_title("Maximum prediction")
ax[2].set_aspect('equal', 'box')
ax[2].set_xticks(np.arange(0,12,2))

fig.colorbar(cax, ax=ax.ravel().tolist())

fig.savefig('kernel_reg.png')
```
